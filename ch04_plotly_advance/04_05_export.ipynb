```python
import plotly.graph_objects as go

fig = go.Figure([go.Scatter(x=[1, 2, 3], y=[5, 3, 2])])
fig.write_image("fig1.svg")
```

```python
from IPython.display import Image

img_bytes = fig.to_image(format="png")
Image(img_bytes)
```
