## 2.4 標準的なインタラクティブ操作

### 2. 4. 1 ホバーツール

```python
import plotly.graph_objects as go

go.Figure(
    [
        go.Bar(
            x=[1, 2, 3], y=[1, 2, 3], hovertext=["A社", "B社", "C社"], name="bar"
        ),
        go.Scatter(x=[1, 2, 3], y=[2, 3, 1], name="line"),
    ],
    layout={'hovermode': 'closest'}
).show()
```

### 2. 4. 2 凡例による要素の選択

```python
import numpy as np

np.random.seed(1)
data0 = np.random.randn(2, 200)
data1 = np.random.randn(2, 200) + 1

go.Figure(
    [
        go.Scatter(x=data0[0], y=data0[1], mode="markers", name="A"),
        go.Scatter(x=data1[0], y=data1[1], mode="markers", name="B"),
    ]
).show()
```
